```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
from pathlib import Path
import os
#import sqlite3
#import json

baseurl = 'https://www.ulta.com/'
headers = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/129.0.0.0 Safari/537.36'
}

productlinks = []
#69
for x in range (1):
    r = requests.get(f'https://www.ulta.com/shop/skin-care/all?page={x}', headers=headers)
    soup = BeautifulSoup(r.content, 'lxml')
    productlist = soup.find_all('li', class_='ProductListingResults__productCard')
    for item in productlist:
        for link in item.find_all('a', href=True):
            productlinks.append(link['href'])

print(productlinks[0])

testlink = productlinks[0]
r = requests.get(testlink, headers = headers)
soup = BeautifulSoup(r.content, 'lxml')

#brand = soup.find('a', {'class' : 'css-11cofee eanm77i0'}).text.strip()
#productName = soup.find('span', {'class' : 'css-1pgnl76 eanm77i0'}).text.strip()

brand = soup.find_all('a', class_='pal-c-Link pal-c-Link--primary pal-c-Link--compact')
name = soup.find('span', class_='Text-ds Text-ds--title-5 Text-ds--left Text-ds--black').text.strip()
ingredients = soup.find('div', class_='Markdown Markdown--body-2 Markdown--left').text.strip()
ingredientss = soup.find_all('div', class_='Markdown Markdown--body-2 Markdown--left')

print(brand)
print(ingredientss)


products_data = []
for link in productlinks:
    r = requests.get(link, headers=headers)
    soup = BeautifulSoup(r.content, 'lxml')

    try:
        # Extracting product name and ingredients
        brand = soup.find_all('a', class_='pal-c-Link pal-c-Link--primary pal-c-Link--compact')
        name = soup.find('span', class_='Text-ds Text-ds--title-5 Text-ds--left Text-ds--black').text.strip()
        ingredients = soup.find('div', class_='Markdown Markdown--body-2 Markdown--left').text.strip()
        ingredientss = soup.find_all('div', class_='Markdown Markdown--body-2 Markdown--left')

        skinCare = {
            'brand': brand,
            'name': name,
            'ingredients' : ingredientss
        }

        # Append the baby food data to the product list
        products_data.append(skinCare)

    except Exception as e:
        print(f"Error while scraping {link}: {e}")

df = pd.DataFrame(products_data)

#filepath = r'C:\Users\moong\Downloads\test.csv'
#df.to_csv(filepath, encoding="utf-8")

testlink = 'https://www.ulta.com/p/mineral-mattescreen-spf-40-pimprod2029941?sku=2589388'
r = requests.get(testlink, headers = headers)
soup = BeautifulSoup(r.content, 'lxml')

brand = soup.find('a', {'class' : 'pal-c-Link pal-c-Link--primary pal-c-Link--compact'}).text.strip()
#<a class="pal-c-Link pal-c-Link--primary pal-c-Link--compact" href="https://www.ulta.com/brand/supergoop" target="_self" iconsize="default">Supergoop!</a>
productName = soup.find('span', {'class' : 'Text-ds Text-ds--title-5 Text-ds--left Text-ds--black'}).text.strip()
brands = soup.find('a', class_='pal-c-Link pal-c-Link--primary pal-c-Link--compact')
paragraphs = brands.find_all('p')

#ingredientsss = soup.find('div', attrs={'id':'ingredients'})
#ingredientss = soup.find('div', class_='css-1ue8dmw eanm77i0')
#ingredients = soup.findAll('div', {'class' : 'css-1ue8dmw eanm77i0'})

print(brand)
print(brands)
print(paragraphs)
#print(productName)
#print(brands)
#print(ingredientss)
#display(ingredients)
```
